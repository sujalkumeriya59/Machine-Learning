```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
# We'll use the 'tips' dataset that comes with seaborn
# This dataset contains information about tips received by waiters and waitresses
data = sns.load_dataset('tips')

# Display the first few rows of the dataset to understand its structure
print("Dataset preview:")
print(data.head())

# Perform a basic analysis
# Analyze the relationship between total bill and tip amounts
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(data=data, x='total_bill', y='tip', hue='time', style='sex', size='size', sizes=(20, 200))

# Customize the plot
plt.title('Scatter Plot of Total Bill vs Tip', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Tip ($)', fontsize=14)
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()

# Save the visualization
plt.savefig("scatter_plot_tips.png")
```
